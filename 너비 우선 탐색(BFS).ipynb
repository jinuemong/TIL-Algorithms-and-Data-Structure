### 너비 우선 탐색 (Breadth-First Search)

- 정점들과 같은 레벨에 있는 노드들 (형제 노드)을 먼저 탐색하는 방식
- 방문 노드   : 큐 사용
- 방문할 노드 : 큐 사용 

####  파이썬에서 딕셔너리를 이용해서 그래프 표현
- 인접 노드를 키 - 값 형태로 구현
- key - value


```python
graph = dict()
graph['A'] = ['B','C']
graph['B'] = ['A','D']
graph['C'] = ['A','G','H','I']
graph['D'] = ['B','E','F']
graph['E'] = ['D']
graph['F'] = ['D']
graph['G'] = ['C']
graph['H'] = ['C']
graph['I'] = ['C','J']
graph['J'] = ['I']
graph
```

```python
def bfs(graph,start_node):
    visited = list() # 방문 완료 큐
    need_visit = list() #방문 할 큐
    
    need_visit.append(start_node)
    
    while need_visit: #need_visit 큐가 존재 할 때까지(방문 할 노드가 남은 경우)
        node = need_visit.pop(0)
        if node not in visited: #파이썬의 장점 - 바로 탐색 (다른 언어는 하나씩 확인)
            #방문 완료 노드에 해당 노드가 없다면 
            visited.append(node) #해당 노드를 방문 완료로 전환
            
            need_visit.extend(graph[node]) #키 값으로 value 전체 추가
            
    return visited
        
```

```python
bfs(graph,'A')
```

#### 시간 복잡도

- 일반적인 BFS 시간 복잡도
- 노드 수 : V , 간선 수: E
- 시간 복잡도: O(V + E)

### 깊이 우선 탐색 (Depth-First Search)

- 정점의 자식들을 먼저 탐색하는 방식
- 방문 노드   : 큐 사용
- 방문할 노드 : 스택 사용

```python
graph = dict()
graph['A'] = ['B','C']
graph['B'] = ['A','D']
graph['C'] = ['A','G','H','I']
graph['D'] = ['B','E','F']
graph['E'] = ['D']
graph['F'] = ['D']
graph['G'] = ['C']
graph['H'] = ['C']
graph['I'] = ['C','J']
graph['J'] = ['I']
graph
```

```python
def dfs(graph,start_node):
    visited = list() # 방문 완료 큐
    need_visit = list() #방문 할 스택
    
    need_visit.append(start_node)
    
    while need_visit: #need_visit 큐가 존재 할 때까지(방문 할 노드가 남은 경우)
        node = need_visit.pop() #스택이니까 뒤에서 부터 빼기 
        if node not in visited: #파이썬의 장점 - 바로 탐색 (다른 언어는 하나씩 확인)
            #방문 완료 노드에 해당 노드가 없다면 
            visited.append(node) #해당 노드를 방문 완료로 전환
            
            need_visit.extend(graph[node]) #키 값으로 value 전체 추가
            
    return visited
```

```python
dfs(graph,'A')
```

#### 시간 복잡도

- 일반적인 BFS 시간 복잡도
- 노드 수 : V , 간선 수: E
- 시간 복잡도: O(V + E)
